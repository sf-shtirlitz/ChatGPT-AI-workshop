```python
from langchain import PromptTemplate

template = '''
		Context: {context}
		Task: {task}
		Format: {expected_format}
'''

prompt = PromptTemplate(
    input_variables=['context','task', 'expected_format'],
    template=template,
)

context = 'In a world transitioning to sustainable energy'
task = 'Explain the benefits and challenges of electric cars'
expected_format = 'Provide a balanced analysis, discussing environmental impact, technology, and adoption barriers'

final_prompt = prompt.format(context=context, task=task, expected_format=expected_format)

print(f'Final Prompt: {final_prompt}')

llm(final_prompt)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage

chat = ChatOpenAI() #

res = chat(
    [
        SystemMessage(
            content="You are Mr Solar System, an expert on Astronomy. You know everything about Solar System. You do NOT know anything about topics other than Solar System. You are truthful and never lie. Never make up facts and if you are not 100% sure, reply with why you cannot answer in a truthful way. Respond in short sentences. Shape your response as if talking to a 10-years-old."
        ),
        HumanMessage(content="Tell me about pluto?")
    ]
)

print(res.content)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage

chat = ChatOpenAI()

res = chat(
    [
        SystemMessage(
            content="You are Mr Solar System, an expert on Astronomy. You know everything about Solar System. You do NOT know anything about topics other than Solar System. You are truthful and never lie. Never make up facts and if you are not 100% sure, reply with why you cannot answer in a truthful way. Respond in short sentences. Shape your response as if talking to a 10-years-old."
        ),
        HumanMessage(content="Tell me about cloud certifications?")
    ]
)

print(res.content)
```
